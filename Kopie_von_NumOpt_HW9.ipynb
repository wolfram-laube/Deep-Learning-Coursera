Problem 1. Exercise 12.19 from the book ((a)-(d)).

### Problem Statement
We are given the following optimization problem:
$$
\min_{x \in \mathbb{R}^2} f(x) = -2x_1 + x_2 \quad \text{subject to} \quad
\begin{cases}
(1 - x_1)^3 - x_2 \geq 0 \\
x_2 + 0.25x_1^2 - 1 \geq 0
\end{cases}
$$
The optimal solution is $ x^* = (0, 1)^T $, where both constraints are active.

#### (a) Do the LICQ hold at this point?

**LICQ (Linear Independence Constraint Qualification):**
LICQ holds if the gradients of the active constraints at $ x^* $ are linearly independent.

First, identify the active constraints at $ x^* $:
- Constraint 1: $(1 - x_1)^3 - x_2 = 0$
- Constraint 2: $x_2 + 0.25x_1^2 - 1 = 0$

At $ x^* = (0, 1)^T $:
- Constraint 1: $(1 - 0)^3 - 1 = 0$
- Constraint 2: $1 + 0.25 \cdot 0^2 - 1 = 0$

Next, compute the gradients of the active constraints at $ x^* $:
- $\nabla g_1(x) = \left( -3(1 - x_1)^2, -1 \right)$
- $\nabla g_2(x) = \left( 0.5 x_1, 1 \right)$

At $ x^* = (0, 1) $:
- $\nabla g_1(x^*) = (-3, -1)$
- $\nabla g_2(x^*) = (0, 1)$

Check for linear independence:
The gradients $\nabla g_1(x^*) = (-3, -1)$ and $\nabla g_2(x^*) = (0, 1)$ are linearly independent because there are no constants $ \alpha $ and $ \beta $ (not both zero) such that $ \alpha (-3, -1) + \beta (0, 1) = (0, 0) $. Hence, LICQ holds at $ x^* $.

#### (b) Are the KKT conditions satisfied?

The KKT conditions are:
1. Stationarity: $\nabla f(x^*) + \lambda_1 \nabla g_1(x^*) + \lambda_2 \nabla g_2(x^*) = 0$
2. Primal feasibility: $g_1(x^*) \geq 0$, $g_2(x^*) \geq 0$
3. Dual feasibility: $\lambda_1 \geq 0$, $\lambda_2 \geq 0$
4. Complementary slackness: $\lambda_1 g_1(x^*) = 0$, $\lambda_2 g_2(x^*) = 0$

Let's check these one by one.

1. **Stationarity:**
   $$
   \nabla f(x) = (-2, 1)
   $$
   $$
   \nabla f(x^*) + \lambda_1 \nabla g_1(x^*) + \lambda_2 \nabla g_2(x^*) = (-2, 1) + \lambda_1 (-3, -1) + \lambda_2 (0, 1) = 0
   $$
   This gives us two equations:
   $$
   -2 - 3\lambda_1 = 0 \quad \Rightarrow \quad \lambda_1 = -\frac{2}{3}
   $$
   $$
   1 - \lambda_1 + \lambda_2 = 0 \quad \Rightarrow \quad 1 + \frac{2}{3} + \lambda_2 = 0 \quad \Rightarrow \quad \lambda_2 = -\frac{5}{3}
   $$

   Since $\lambda_1$ and $\lambda_2$ must be non-negative (dual feasibility), these values are not valid. Therefore, the KKT conditions are not satisfied.

#### (c) Write down the sets $\mathcal{F}(x^*)$ and $\mathcal{C}(x^*, \lambda^*)$.

The sets are:
- $\mathcal{F}(x^*)$ is the set of active constraint indices at $ x^* $:
  $$
  \mathcal{F}(x^*) = \{1, 2\}
  $$
- $\mathcal{C}(x^*, \lambda^*)$ is the critical cone, consisting of all $d$ such that:
  $$
  \nabla f(x^*)^T d \leq 0 \quad \text{and} \quad \nabla g_i(x^*)^T d = 0 \quad \forall i \in \mathcal{F}(x^*)
  $$

Given the violation of KKT conditions, we can't form $\mathcal{C}(x^*, \lambda^*)$ accurately.

#### (d) Are the second-order necessary conditions satisfied? Are the second-order sufficient conditions satisfied?

Second-order necessary conditions require that:
$$
d^T \nabla^2 L(x^*, \lambda^*) d \geq 0 \quad \forall d \in \mathcal{C}(x^*, \lambda^*)
$$
where $\nabla^2 L(x, \lambda) = \nabla^2 f(x) + \sum_{i \in \mathcal{F}(x)} \lambda_i \nabla^2 g_i(x)$.

Given that the first-order conditions (KKT) are not satisfied, second-order conditions are not relevant for this analysis, and therefore we can't conclude about them.

In conclusion:
- LICQ holds.
- KKT conditions are not satisfied.
- The sets $\mathcal{F}(x^*)$ and $\mathcal{C}(x^*, \lambda^*)$ are partially determined.
- Second-order conditions can't be verified due to the failure of KKT conditions.


Problem 2. Exercise 12.20 from the book (minima over the unit circle). Use second-order conditions to justify your answer.

### Problem Statement
We need to find the minima of the function $ f(x) = x_1 x_2 $ on the unit circle $ x_1^2 + x_2^2 = 1 $ and illustrate this problem geometrically.

### Solution
To find the extrema of $ f(x) = x_1 x_2 $ on the unit circle, we can use the method of Lagrange multipliers. We introduce a Lagrange multiplier $ \lambda $ and define the Lagrangian function:
$$
\mathcal{L}(x_1, x_2, \lambda) = x_1 x_2 + \lambda (x_1^2 + x_2^2 - 1)
$$

#### Step 1: Compute the gradients
We need to set the gradients of $ \mathcal{L} $ to zero:
$$
\nabla \mathcal{L} = \left( \frac{\partial \mathcal{L}}{\partial x_1}, \frac{\partial \mathcal{L}}{\partial x_2}, \frac{\partial \mathcal{L}}{\partial \lambda} \right) = 0
$$

The partial derivatives are:
$$
\frac{\partial \mathcal{L}}{\partial x_1} = x_2 + 2\lambda x_1 = 0
$$
$$
\frac{\partial \mathcal{L}}{\partial x_2} = x_1 + 2\lambda x_2 = 0
$$
$$
\frac{\partial \mathcal{L}}{\partial \lambda} = x_1^2 + x_2^2 - 1 = 0
$$

#### Step 2: Solve the system of equations
From the first two equations, we get:
$$
x_2 + 2\lambda x_1 = 0 \quad \Rightarrow \quad x_2 = -2\lambda x_1
$$
$$
x_1 + 2\lambda x_2 = 0 \quad \Rightarrow \quad x_1 = -2\lambda x_2
$$

Substituting $ x_2 = -2\lambda x_1 $ into $ x_1 = -2\lambda x_2 $, we get:
$$
x_1 = -2\lambda (-2\lambda x_1) \quad \Rightarrow \quad x_1 = 4\lambda^2 x_1
$$

For non-trivial solutions ($ x_1 \neq 0 $), we get:
$$
4\lambda^2 = 1 \quad \Rightarrow \quad \lambda = \pm \frac{1}{2}
$$

#### Step 3: Determine the critical points
For $ \lambda = \frac{1}{2} $:
$$
x_2 = -2\left(\frac{1}{2}\right) x_1 = -x_1
$$
Substitute into the constraint $ x_1^2 + x_2^2 = 1 $:
$$
x_1^2 + (-x_1)^2 = 1 \quad \Rightarrow \quad 2x_1^2 = 1 \quad \Rightarrow \quad x_1^2 = \frac{1}{2} \quad \Rightarrow \quad x_1 = \pm \frac{1}{\sqrt{2}}
$$
Thus, the corresponding $ x_2 $ values are:
$$
x_2 = \mp \frac{1}{\sqrt{2}}
$$

The critical points are $ \left( \frac{1}{\sqrt{2}}, -\frac{1}{\sqrt{2}} \right) $ and $ \left( -\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}} \right) $.

For $ \lambda = -\frac{1}{2} $:
$$
x_2 = -2\left(-\frac{1}{2}\right) x_1 = x_1
$$
Substitute into the constraint $ x_1^2 + x_2^2 = 1 $:
$$
x_1^2 + x_1^2 = 1 \quad \Rightarrow \quad 2x_1^2 = 1 \quad \Rightarrow \quad x_1^2 = \frac{1}{2} \quad \Rightarrow \quad x_1 = \pm \frac{1}{\sqrt{2}}
$$
Thus, the corresponding $ x_2 $ values are:
$$
x_2 = \pm \frac{1}{\sqrt{2}}
$$

The critical points are $ \left( \frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}} \right) $ and $ \left( -\frac{1}{\sqrt{2}}, -\frac{1}{\sqrt{2}} \right) $.

#### Step 4: Evaluate $ f(x) $ at critical points
$$
f\left( \frac{1}{\sqrt{2}}, -\frac{1}{\sqrt{2}} \right) = \frac{1}{\sqrt{2}} \cdot -\frac{1}{\sqrt{2}} = -\frac{1}{2}
$$
$$
f\left( -\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}} \right) = -\frac{1}{\sqrt{2}} \cdot \frac{1}{\sqrt{2}} = -\frac{1}{2}
$$
$$
f\left( \frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}} \right) = \frac{1}{\sqrt{2}} \cdot \frac{1}{\sqrt{2}} = \frac{1}{2}
$$
$$
f\left( -\frac{1}{\sqrt{2}}, -\frac{1}{\sqrt{2}} \right) = -\frac{1}{\sqrt{2}} \cdot -\frac{1}{\sqrt{2}} = \frac{1}{2}
$$

Thus, the minima of the function are $ -\frac{1}{2} $ at the points $ \left( \frac{1}{\sqrt{2}}, -\frac{1}{\sqrt{2}} \right) $ and $ \left( -\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}} \right) $.

### Geometric Illustration
To illustrate this problem geometrically, we will plot the unit circle $ x_1^2 + x_2^2 = 1 $ and indicate the points where the function $ f(x) = x_1 x_2 $ achieves its minimum values.

#### Generating the plot
Let's create the plot to visualize this.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the circle
theta = np.linspace(0, 2 * np.pi, 300)
x1 = np.cos(theta)
x2 = np.sin(theta)

# Critical points
critical_points = [
    (1/np.sqrt(2), -1/np.sqrt(2)),
    (-1/np.sqrt(2), 1/np.sqrt(2)),
    (1/np.sqrt(2), 1/np.sqrt(2)),
    (-1/np.sqrt(2), -1/np.sqrt(2))
]

# Plot the circle
plt.figure(figsize=(6, 6))
plt.plot(x1, x2, label='$x_1^2 + x_2^2 = 1$')
plt.scatter(*zip(*critical_points), color='red', zorder=5)
plt.text(1/np.sqrt(2), -1/np.sqrt(2), 'Minima', color='red', ha='right')
plt.text(-1/np.sqrt(2), 1/np.sqrt(2), 'Minima', color='red', ha='left')
plt.text(1/np.sqrt(2), 1/np.sqrt(2), 'Maxima', color='blue', ha='left')
plt.text(-1/np.sqrt(2), -1/np.sqrt(2), 'Maxima', color='blue', ha='right')

# Labels and title
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title('Geometric Illustration of $f(x) = x_1 x_2$ on the Unit Circle')
plt.show()
```


#### Explanation
- The unit circle $ x_1^2 + x_2^2 = 1 $ is plotted.
- The minima of the function $ f(x) = x_1 x_2 $ occur at the points $ \left( \frac{1}{\sqrt{2}}, -\frac{1}{\sqrt{2}} \right) $ and $ \left( -\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}} \right) $, marked in red.
-

 The maxima of the function $ f(x) = x_1 x_2 $ occur at the points $ \left( \frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}} \right) $ and $ \left( -\frac{1}{\sqrt{2}}, -\frac{1}{\sqrt{2}} \right) $, marked in blue.

Problem 3. Exercise 12.21 from the book (maxima over the unit disk). Use second-order conditions to justify your answer.

### Problem Statement
We need to find the maxima of the function $ f(x) = x_1 x_2 $ over the unit disk defined by the inequality constraint $ 1 - x_1^2 - x_2^2 \geq 0 $.

### Solution
To find the extrema of $ f(x) = x_1 x_2 $ on the unit disk, we consider the boundary of the disk $ x_1^2 + x_2^2 = 1 $ and the interior of the disk $ x_1^2 + x_2^2 < 1 $.

#### Boundary Analysis
On the boundary of the disk $ x_1^2 + x_2^2 = 1 $, we have already determined the critical points in the previous problem. The maxima of $ f(x) $ on the boundary are:
$$
\left( \frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}} \right) \quad \text{and} \quad \left( -\frac{1}{\sqrt{2}}, -\frac{1}{\sqrt{2}} \right)
$$
with function value:
$$
f\left( \frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}} \right) = \frac{1}{2}
$$
$$
f\left( -\frac{1}{\sqrt{2}}, -\frac{1}{\sqrt{2}} \right) = \frac{1}{2}
$$

#### Interior Analysis
Within the interior of the disk $ x_1^2 + x_2^2 < 1 $, we need to check if there are any stationary points of $ f(x) = x_1 x_2 $.

The first-order partial derivatives of $ f(x) $ are:
$$
\frac{\partial f}{\partial x_1} = x_2
$$
$$
\frac{\partial f}{\partial x_2} = x_1
$$

Setting these derivatives to zero:
$$
x_2 = 0 \quad \text{and} \quad x_1 = 0
$$

Thus, the only stationary point within the interior of the disk is at $ (0, 0) $, where $ f(0, 0) = 0 $.

#### Conclusion
Since the maximum value of $ f(x) = x_1 x_2 $ on the boundary of the unit disk is $ \frac{1}{2} $ and the value at the interior stationary point is $ 0 $, the maximum of $ f(x) $ over the entire unit disk occurs on the boundary.

### Geometric Illustration
To illustrate this problem geometrically, we will plot the unit disk and indicate the points where the function $ f(x) = x_1 x_2 $ achieves its maximum values.

#### Generating the plot
Let's create the plot to visualize this.



```python
import numpy as np
import matplotlib.pyplot as plt

# Define the circle
theta = np.linspace(0, 2 * np.pi, 300)
x1 = np.cos(theta)
x2 = np.sin(theta)

# Critical points on the boundary
boundary_max_points = [
    (1/np.sqrt(2), 1/np.sqrt(2)),
    (-1/np.sqrt(2), -1/np.sqrt(2))
]

# Plot the circle and the critical points
plt.figure(figsize=(6, 6))
plt.fill(x1, x2, alpha=0.2, label='Unit Disk $x_1^2 + x_2^2 \\leq 1$')
plt.scatter(*zip(*boundary_max_points), color='red', zorder=5, label='Boundary Maxima')
plt.scatter(0, 0, color='blue', zorder=5, label='Interior Stationary Point')

# Labels and title
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend(loc='upper right')
plt.title('Geometric Illustration of $f(x) = x_1 x_2$ over the Unit Disk')
plt.show()
```


This plot illustrates:
- The unit disk $ x_1^2 + x_2^2 \leq 1 $.
- The points $ \left( \frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}} \right) $ and $ \left( -\frac{1}{\sqrt{2}}, -\frac{1}{\sqrt{2}} \right) $ where the function $ f(x) = x_1 x_2 $ achieves its maximum value of $ \frac{1}{2} $.
- The interior stationary point $ (0, 0) $ with function value $ 0 $.

The contour plot of the function $ f(x) = x_1 x_2 $ over the unit disk is shown below. The unit disk is shaded, and the contour lines represent the values of $ f(x) $. The boundary maxima are indicated in red, and the interior stationary point is marked in blue. This visualization highlights how the function varies over the unit disk and clearly shows the points where the function reaches its maximum and stationary values.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and the circle
def f(x1, x2):
    return x1 * x2

x = np.linspace(-1, 1, 300)
y = np.linspace(-1, 1, 300)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Plot the circle and the contour lines
plt.figure(figsize=(6, 6))
plt.fill(x1, x2, alpha=0.2, label='Unit Disk $x_1^2 + x_2^2 \\leq 1$')
plt.contour(X, Y, Z, levels=20, cmap='coolwarm')
plt.scatter(*zip(*boundary_max_points), color='red', zorder=5, label='Boundary Maxima')
plt.scatter(0, 0, color='blue', zorder=5, label='Interior Stationary Point')

# Labels and title
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend(loc='upper right')
plt.title('Contour Plot of $f(x) = x_1 x_2$ over the Unit Disk')
plt.show()
```
